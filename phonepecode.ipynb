```python
import os
import json
import pandas as pd
import mysql.connector
```

```python
#Collecting data from aggregated_insurance

path1="K:/DS/pulse/data/aggregated/insurance/country/india/state/"

aggregated_insurance_list= os.listdir(path1)

columns1={"States":[], "Years":[], "Quarters":[], "Trans_type":[], "Trans_count":[], "Trans_amount":[]}

for state in aggregated_insurance_list:
    current_states=path1+state+"/" 
    # print(current_states)
    aggregated_year_list=os.listdir(current_states)
    # print(aggregated_year_list)

    for year in aggregated_year_list:
        current_year=current_states+year+"/"
        # print(current_year)
        aggregated_quarts_list=os.listdir(current_year)
        # print(aggregated_quarts_list)

        for quarts in aggregated_quarts_list:
            current_quarts=current_year+quarts
            data=open(current_quarts,"r")

            A = json.load(data)
            
            for i in A["data"]["transactionData"]:
                # print(i)
                name=i["name"]
                # print(name)
                count=i["paymentInstruments"][0]["count"]
                amount=i["paymentInstruments"][0]["amount"]
                columns1["Trans_type"].append(name)
                columns1["Trans_count"].append(count)
                columns1["Trans_amount"].append(amount)
                columns1["States"].append(state)
                columns1["Years"].append(year)
                columns1["Quarters"].append(int(quarts.strip(".json")))

aggregated_insurance=pd.DataFrame(columns1)

aggregated_insurance["States"]=aggregated_insurance["States"].str.replace('andaman-&-nicobar-islands', 'Andaman & Nicobar')
aggregated_insurance["States"]=aggregated_insurance["States"].str.replace("-"," ")
aggregated_insurance["States"]=aggregated_insurance["States"].str.title()
aggregated_insurance["States"]=aggregated_insurance["States"].str.replace('Dadra & Nagar Haveli & Daman & Diu','Dadra and Nagar Haveli and Daman and Diu')
```

```python
#Collecting data from aggregated_user

path3="K:/DS/pulse/data/aggregated/user/country/india/state/"

aggregated_user_list= os.listdir(path3)

columns3={"States":[], "Years":[], "Quarters":[], "Brand":[], "Count":[], "Percentage":[]}

for state in aggregated_user_list:
    current_states=path3+state+"/" 
    # print(current_states)
    aggregated_year_list=os.listdir(current_states)
    # print(aggregated_year_list)

    for year in aggregated_year_list:
        current_year=current_states+year+"/"
        #print(current_year)
        aggregated_quarts_list=os.listdir(current_year)
        #print(aggregated_quarts_list)

        for quarts in aggregated_quarts_list:
            current_quarts=current_year+quarts
            data=open(current_quarts,"r")

            C = json.load(data)
            # print(C["data"]["usersByDevice"])

            try:

                for i in C["data"]["usersByDevice"]:
                    #print(i)
                    brand=i["brand"]
                    # print(name)
                    count=i["count"]
                    percentage=i["percentage"]
                    columns3["Brand"].append(brand)
                    columns3["Count"].append(count)
                    columns3["Percentage"].append(percentage)
                    columns3["States"].append(state)
                    columns3["Years"].append(year)
                    columns3["Quarters"].append(int(quarts.strip(".json")))
            
            except:
                pass

aggregated_user=pd.DataFrame(columns3)

aggregated_user["States"]=aggregated_user["States"].str.replace('andaman-&-nicobar-islands', 'Andaman & Nicobar')
aggregated_user["States"]=aggregated_user["States"].str.replace("-"," ")
aggregated_user["States"]=aggregated_user["States"].str.title()
aggregated_user["States"]=aggregated_user["States"].str.replace('Dadra & Nagar Haveli & Daman & Diu','Dadra and Nagar Haveli and Daman and Diu')
```
